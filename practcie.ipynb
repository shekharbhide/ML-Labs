```python
import pandas as pd
data = pd.read_csv('Mall_Customers.csv')
```

```python
# extractigrelevant feautrs
x = data.iloc[:, [3,4]].values
```

```python
# traing algo
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5, init='k-means++',random_state=3)
```

```python
clusters = kmeans.fit_predict(x)
```

```python
from matplotlib import pyplot as plt
plt.scatter(x[clusters==0,0],x[clusters==0,1],s=100,c='blue',label='cluster1')
plt.scatter(x[clusters==1,0],x[clusters==1,1],s=100,c='red',label='cluster2')
plt.scatter(x[clusters==2,0],x[clusters==2,1],s=100,c='cyan',label='clsuster3')
plt.scatter(x[clusters==3,0],x[clusters==3,1],s=100,c='brown',label='cluster4')
plt.scatter(x[clusters==4,0],x[clusters==4,1],s=100,c='black',label='cluster5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='centroid')

plt.title('cluster ofcustomer')
plt.xlabel('annaul income')
plt.ylabel('speed')
plt.legend()
plt.show()
```
